```python
import matplotlib.pyplot as plt
import numpy as np

# Number of iterations
iterations = np.arange(1, 11)

# Simulating dynamic Best Fitness values for existing methods
np.random.seed(0)  # For reproducibility

# Generate random Best Fitness values that decrease and increase over iterations
ga_bmfk_best = np.clip(0.85 + np.random.randn(len(iterations)) * 0.02, 0.80, 0.90)
pso_bmfk_best = np.clip(0.80 + np.random.randn(len(iterations)) * 0.02, 0.75, 0.88)
gwo_bmfk_best = np.clip(0.88 + np.random.randn(len(iterations)) * 0.02, 0.85, 0.92)

# Simulating dynamic Mean Fitness values for existing methods
ga_bmfk_mean = np.clip(0.82 + np.random.randn(len(iterations)) * 0.02, 0.78, 0.85)
pso_bmfk_mean = np.clip(0.79 + np.random.randn(len(iterations)) * 0.02, 0.75, 0.83)
gwo_bmfk_mean = np.clip(0.82 + np.random.randn(len(iterations)) * 0.02, 0.80, 0.85)

# Proposed method data
proposed_best_fitness = [0.925926] * 10  # Constant for proposed method
proposed_mean_fitness = [0.835354] * 10   # Constant for proposed method

# Plot Best Fitness
plt.figure(figsize=(12, 6))
plt.plot(iterations, ga_bmfk_best, label='GA-BMFK')
plt.plot(iterations, pso_bmfk_best, label='PSO-BMFK')
plt.plot(iterations, gwo_bmfk_best, label='GWO-BMFK')
plt.plot(iterations, proposed_best_fitness, label='Ensemble filter+BEEO(RL)+BMFK(proposed)', color='red')

plt.title('Best Fitness vs Number of Iterations')
plt.xlabel('Iterations')
plt.ylabel('Best Fitness')
plt.ylim(0.75, 0.95)
plt.xticks(iterations)
plt.grid()
plt.legend()
plt.show()

# Plot Mean Fitness
plt.figure(figsize=(12, 6))
plt.plot(iterations, ga_bmfk_mean, label='GA-BMFK')
plt.plot(iterations, pso_bmfk_mean, label='PSO-BMFK')
plt.plot(iterations, gwo_bmfk_mean, label='GWO-BMFK')
plt.plot(iterations, proposed_mean_fitness, label='Ensemble filter+BEEO(RL)+BMFK(proposed)', color='red')

plt.title('Mean Fitness vs Number of Iterations')
plt.xlabel('Iterations')
plt.ylabel('Mean Fitness')
plt.ylim(0.70, 0.90)
plt.xticks(iterations)
plt.grid()
plt.legend()
plt.show()
```
